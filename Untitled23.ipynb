```python
X = df.drop("target_binary", axis=1)
y = df["target_binary"]

print("Features Shape:", X.shape)
print("Target Shape:", y.shape)
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)

IQR = Q3 - Q1

df = df[
    ~((df < (Q1 - 1.5 * IQR)) |
      (df > (Q3 + 1.5 * IQR))).any(axis=1)
]

print("Dataset Shape After Removing Outliers:")
print(df.shape)
```

```python
X = df.drop(["num", "target_binary"], axis=1)
y = df["target_binary"]

print(X.columns)
print(X.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print("Scaling Completed")
```

```python
from sklearn.feature_selection import SelectKBest, f_classif

selector = SelectKBest(score_func=f_classif, k=10)

X_selected = selector.fit_transform(X_scaled, y)

print("Original Shape:", X_scaled.shape)
print("Selected Shape:", X_selected.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_selected,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("Training Shape:", X_train.shape)
print("Testing Shape:", X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(random_state=42)

lr.fit(X_train, y_train)

lr_predictions = lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

print("Accuracy:", accuracy_score(y_test, lr_predictions))
print("Precision:", precision_score(y_test, lr_predictions))
print("Recall:", recall_score(y_test, lr_predictions))
print("F1 Score:", f1_score(y_test, lr_predictions))

print("\nConfusion Matrix")
print(confusion_matrix(y_test, lr_predictions))
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=42)

dt.fit(X_train, y_train)

dt_predictions = dt.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, dt_predictions))
print("Precision:", precision_score(y_test, dt_predictions))
print("Recall:", recall_score(y_test, dt_predictions))
print("F1 Score:", f1_score(y_test, dt_predictions))

print("\nConfusion Matrix")
print(confusion_matrix(y_test, dt_predictions))
```

```python
overfit_model = DecisionTreeClassifier(
    max_depth=None,
    random_state=42
)

overfit_model.fit(X_train, y_train)

train_pred = overfit_model.predict(X_train)
test_pred = overfit_model.predict(X_test)

print("Training Accuracy:", accuracy_score(y_train, train_pred))
print("Testing Accuracy:", accuracy_score(y_test, test_pred))
```

```python
pruned_model = DecisionTreeClassifier(
    max_depth=3,
    random_state=42
)

pruned_model.fit(X_train, y_train)

pruned_pred = pruned_model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, pruned_pred))
print("Precision:", precision_score(y_test, pruned_pred))
print("Recall:", recall_score(y_test, pruned_pred))
print("F1 Score:", f1_score(y_test, pruned_pred))

print("\nConfusion Matrix")
print(confusion_matrix(y_test, pruned_pred))
```

```python
underfit_model = DecisionTreeClassifier(
    max_depth=1,
    random_state=42
)

underfit_model.fit(X_train, y_train)

underfit_pred = underfit_model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, underfit_pred))
print("Precision:", precision_score(y_test, underfit_pred))
print("Recall:", recall_score(y_test, underfit_pred))
print("F1 Score:", f1_score(y_test, underfit_pred))
```

```python
improved_model = DecisionTreeClassifier(
    max_depth=5,
    random_state=42
)

improved_model.fit(X_train, y_train)

improved_pred = improved_model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, improved_pred))
print("Precision:", precision_score(y_test, improved_pred))
print("Recall:", recall_score(y_test, improved_pred))
print("F1 Score:", f1_score(y_test, improved_pred))

print("\nConfusion Matrix")
print(confusion_matrix(y_test, improved_pred))
```

```python
print(" Model Comparison ")
print(f"Logistic Regression Accuracy: {accuracy_score(y_test, lr_predictions):.4f}")
print(f"Decision Tree Accuracy: {accuracy_score(y_test, dt_predictions):.4f}")
print(f"Pruned Decision Tree Accuracy: {accuracy_score(y_test, pruned_pred):.4f}")
print(f"Improved Decision Tree Accuracy: {accuracy_score(y_test, improved_pred):.4f}")
```
